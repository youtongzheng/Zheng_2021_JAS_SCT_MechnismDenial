```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from matplotlib import cm
```

```python
f = xr.open_mfdataset('./*res.nc',
                      combine = 'by_coords',decode_cf = False)
```

```python
myfontsize = 12
```

```python
# defining a function for axe plotting
def pltbflx(iax, it, iz, ivar, ivarzi, ivarzb, ivarBIR,
           ylabel = 'z (m)'):
    mylevels = np.linspace(-25, 25, 13)

    oax = iax.contourf(it,iz,ivar, cmap = cm.RdBu_r, levels = mylevels, extend = 'max')

    iax.plot(it, ivarzi, color = 'black',label=r"$\mathrm{z_i}$")
    iax.plot(it, ivarzb, color = 'black',label=r"$\mathrm{z_b}$", linestyle = '--')

    iax.set_ylabel(ylabel, fontsize=myfontsize)  # Add an x-label to the axes.
    iax.set_ylim([0., 1500])
    iax.legend(loc="upper left", fontsize=myfontsize, frameon=False, labelspacing = 0.2)

    iax0 = iax.twinx()  # instantiate a second axes that shares the same x-axis
    color = 'tab:blue'
    iax0.set_ylabel('BIR', color=color)  # we already handled the x-label with ax1
    iax0.plot(it, ivarBIR, color=color)
    iax0.tick_params(axis='y', labelcolor=color)
    iax0.set_ylim([-0.01, 0.16])
    
    return oax
```

```python
def plt2var(iax, it, ivar, ivar1, ivarname, ivarname1):
    iax.plot(it, ivar, color = 'black',label=ivarname)
    iax.set_ylabel(ivarname)
    
    iax0 = iax.twinx()  # instantiate a second axes that shares the same x-axis
    color = 'tab:red'
    iax0.set_ylabel(ivarname1, color=color)  # we already handled the x-label with ax1
    iax0.plot(it, ivar1, color=color)
    iax0.tick_params(axis='y', labelcolor=color)
    
    iax.grid(True)
```

```python
def pltcleb(iax, it, ivarEnt, ivarDiab, ivarRes, ivarStore, ivarRad, ivarPrec,
           legend = False):
    iax.plot(it, ivarEnt,label='Ent')
    iax.plot(it, ivarDiab,label='Diab', color = 'black')
    iax.plot(it, ivarRes,label='Base')
    iax.plot(it, ivarStore,label='Stor')
    iax.plot(it, ivarRad,label='Rad', color = 'black', linestyle = '--', alpha=1)
    iax.plot(it, ivarPrec,label='Prec', color = 'black', linestyle = '-.', alpha=1)
    
    iax.set_ylabel(r"$\mathrm{W m^{-2}}$", fontsize=myfontsize)
    iax.set_xlabel("Hour (h)", fontsize=myfontsize)
    iax.set_ylim((-70, 70))
    iax.axhline(y=0., linewidth=1, color='black', linestyle='--')
    if legend:
        iax.legend(bbox_to_anchor=(1.,0.5), loc="center left", fontsize=0.9*myfontsize)

    iax.grid(True)
```

```python
fig, axs = plt.subplots(ncols = 2,
                        nrows = 4,
                        sharex = True, sharey = False,
                        figsize=(18/2.54, (14)/2.54), dpi = 300)
fig.tight_layout(w_pad = 4.5) 

#
oax = pltbflx(axs[0,0], f['t_hr'].where(f['BIR'].isel(x = 2) < 0.15), f['z'], 
              f['sv_flux'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15), 
              f['zi'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15), 
              f['zb'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
              f['BIR'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15))

oax = pltbflx(axs[0,1], f['t_hr'].where(f['BIR'].isel(x = 1) < 0.15), f['z'], 
              f['sv_flux'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15), 
              f['zi'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15), 
              f['zb'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
              f['BIR'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15), ylabel = '')

#
plt2var(axs[1,0], f['t_hr'].where(f['BIR'].isel(x = 2) < 0.15),
        f['LHF'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
        f['Bfx_mean'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15), 
        r"$\mathrm{LHF \ (Wm^{-2})}$", r"$\mathrm{tot. buoy. \ (Wm^{-2})}$")

plt2var(axs[1,1], f['t_hr'].where(f['BIR'].isel(x = 1) < 0.15),
        f['LHF'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
        f['Bfx_mean'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15), 
        r"$\mathrm{LHF \ (Wm^{-2})}$", r"$\mathrm{tot. buoy. \ (Wm^{-2})}$")

#
axs[2,0].plot(f['t_hr'].where(f['BIR'].isel(x = 2) < 0.15),
              f['A'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
              color = 'black',label=r"$\mathrm{A}$")

axs[2,0].plot(f['t_hr'].where(f['BIR'].isel(x = 0) < 0.15),
              f['A'].isel(x = 0).where(f['BIR'].isel(x = 0) < 0.15),
              linestyle = '--',color = 'grey', label=r"$\mathrm{CTRL}$")

axs[2,1].plot(f['t_hr'].where(f['BIR'].isel(x = 1) < 0.15),
              f['A'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
              color = 'black',label=r"$\mathrm{A}$")

axs[2,1].plot(f['t_hr'].where(f['BIR'].isel(x = 0) < 0.15),
              f['A'].isel(x = 0).where(f['BIR'].isel(x = 0) < 0.15),
              linestyle = '--',color = 'grey', label=r"$\mathrm{CTRL}$")

axs[2,0].set_ylabel(r"$\mathrm{A}$")
axs[2,0].set_ylim((0.9, 2.5))
axs[2,0].grid(True)
axs[2,1].set_ylim((0.9, 2.5))
axs[2,1].grid(True)

#
pltcleb(axs[3,0], f['t_hr'].where(f['BIR'].isel(x = 2) < 0.15),
        f['Ent_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15), 
        f['Diab_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
        -f['Res_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
        f['Store_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
        f['Rad_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15),
        f['Prec_CLEB'].isel(x = 2).where(f['BIR'].isel(x = 2) < 0.15))

pltcleb(axs[3,1], f['t_hr'].where(f['BIR'].isel(x = 1) < 0.15),
        f['Ent_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15), 
        f['Diab_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
        -f['Res_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
        f['Store_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
        f['Rad_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
        f['Prec_CLEB'].isel(x = 1).where(f['BIR'].isel(x = 1) < 0.15),
       legend = True)


axs[0,0].set_title('FXDA')
axs[0,1].set_title('FXDLHF')

mylabel = np.array((['a','b'],['c','d'], ['e','f'],['g','h']))

for i in range(4):
    for j in range (2):
        axs[i,j].text(0.04, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,
                      fontsize=myfontsize, fontweight='bold')

fig.savefig('Fig5_MLM.png', dpi=fig.dpi, bbox_inches='tight')
```
