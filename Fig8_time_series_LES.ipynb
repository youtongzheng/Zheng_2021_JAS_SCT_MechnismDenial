```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from matplotlib import cm
from scipy.signal import savgol_filter
```

```python
#read data
path="/lustre/ytzheng/"

Casename = 'ASTEX_cold2.6'
Casename1 = 'ASTEX_cold2.6_fxdlhf'
Casename2 = 'ASTEX_cold2.6_strginv'
Casename3 = 'ASTEX_cold2.6_fxdlhf_strginv'

f = xr.open_mfdataset(path + Casename + '/' + '*N100.nc',
                      combine = 'by_coords',decode_cf = False)
f1 = xr.open_mfdataset(path + Casename1 + '/' + '*N100*.nc',
                      combine = 'by_coords',decode_cf = False)
f2 = xr.open_mfdataset(path + Casename2 + '/' + '*N100.nc',
                      combine = 'by_coords',decode_cf = False)
f3 = xr.open_mfdataset(path + Casename3 + '/' + '*N100*.nc',
                      combine = 'by_coords',decode_cf = False)
```

```python
nt = 90
t_hours = np.arange(nt)

f = f.where(f.CLDLOW > 0.5, drop = True)
f1 = f1.where(f1.CLDLOW > 0.5, drop = True)
f2 = f2.where(f2.CLDLOW > 0.5, drop = True)
f3 = f3.where(f3.CLDLOW > 0.5, drop = True)

myfontsize = 12
```

```python
# defining a function for axe plotting
def plt4l(iax, ix, iy, iy1, iy2, iy3, iytitle, smooth):
    if smooth == 0:
        iax.plot(np.arange(iy.time.size), iy, label = 'CTRL', color = 'crimson')
        iax.plot(np.arange(iy1.time.size), iy1, label = 'FXDLHF', linestyle = '--',color = 'crimson')
        iax.plot(np.arange(iy2.time.size), iy2, label = 'STRGINV', color = 'dodgerblue')
        iax.plot(np.arange(iy3.time.size), iy3, label = 'STRGINV FXDLHF', linestyle = '--', color = 'dodgerblue')
    
    else:
        iax.plot(np.arange(iy.time.size), iy, label = 'CTRL', color = 'crimson', alpha=0.2)
        iax.plot(np.arange(iy1.time.size), iy1, label = 'FXDLHF', linestyle = '--',color = 'crimson', alpha=0.2)
        iax.plot(np.arange(iy2.time.size), iy2, label = 'STRGINV', color = 'dodgerblue', alpha=0.2)
        iax.plot(np.arange(iy3.time.size), iy3, label = 'STRGINV FXDLHF', linestyle = '--', color = 'dodgerblue', alpha=0.2)
        
        yhat = savgol_filter(iy, 21, 3)
        yhat1 = savgol_filter(iy1, 21, 3)
        yhat2 = savgol_filter(iy2, 21, 3)
        yhat3 = savgol_filter(iy3, 21, 3)
        
        iax.plot(np.arange(iy.time.size), yhat, label = 'CTRL', color = 'crimson')
        iax.plot(np.arange(iy1.time.size), yhat1, label = 'FXDLHF', linestyle = '--',color = 'crimson')
        iax.plot(np.arange(iy2.time.size), yhat2, label = 'STRGINV', color = 'dodgerblue')
        iax.plot(np.arange(iy3.time.size), yhat3, label = 'STRGINV FXDLHF', linestyle = '--', color = 'dodgerblue')
        
    iax.set_xlim([0., 60])
    iax.set_ylabel(iytitle, fontsize=myfontsize)  # Add a y-label to the axes.
    iax.grid(True)
```

```python
fig, axs = plt.subplots(ncols = 2,
                        nrows = 4,
                        sharex = True, sharey = False,
                        figsize=(18/2.54, (16)/2.54), dpi = 300)

plt.tight_layout()

#fig a
plt4l(axs[0,0], t_hours, f.TVFLX_we_Wm2_out + f.Prec_wm2_out - f.CldRCool_out, 
      f1.TVFLX_we_Wm2_out + f1.Prec_wm2_out - f1.CldRCool_out, 
      f2.TVFLX_we_Wm2_out + f2.Prec_wm2_out - f2.CldRCool_out,
      f3.TVFLX_we_Wm2_out + f3.Prec_wm2_out - f3.CldRCool_out,
      r"$\mathrm{Ent + Diab \ (Wm^{-2})}$", 1)
axs[0,0].set_ylim((-20., 60))

#fig b
plt4l(axs[0,1], t_hours, f.CldRCool_out, 
      f1.CldRCool_out, 
      f2.CldRCool_out,
      f3.CldRCool_out,
      r"$\mathrm{Rad \ (Wm^{-2})}$", 0)
axs[0,1].legend(loc="lower center", fontsize=0.7*myfontsize)

#fig c
plt4l(axs[1,0], t_hours, f.TVFLX_we_Wm2_out, 
      f1.TVFLX_we_Wm2_out, 
      f2.TVFLX_we_Wm2_out,
      f3.TVFLX_we_Wm2_out,
      r"$\mathrm{Ent \ (Wm^{-2})}$", 1)
axs[1,0].set_ylim((0, 120))

#fig d
plt4l(axs[1,1], t_hours, f.Prec_wm2_out, 
      f1.Prec_wm2_out, 
      f2.Prec_wm2_out,
      f3.Prec_wm2_out,
      r"$\mathrm{Prec \ (Wm^{-2})}$", 1)

#fig e
plt4l(axs[2,0], t_hours, f.delta_thetal_out, 
      f1.delta_thetal_out, 
      f2.delta_thetal_out,
      f3.delta_thetal_out,
      r"$\mathrm{\Delta_{BL} \vartheta_l \ (K)}$", 0)

#fig f
plt4l(axs[2,1], t_hours, 100.*f.we, 
      100.*f1.we, 
      100.*f2.we,
      100.*f3.we,
      r"$\mathrm{w_e \ (cms^{-1})}$", 1)
axs[2,1].set_ylim((0., 2.))

#fig g
plt4l(axs[3,0], t_hours, f.delta_thetal_inv_out, 
      f1.delta_thetal_inv_out, 
      f2.delta_thetal_inv_out,
      f3.delta_thetal_inv_out,
      r"$\mathrm{\Delta_{inv} \vartheta_l \ (K)}$", 0)

#fig h
plt4l(axs[3,1], t_hours, -f.we*f.delta_b_inv_out/f.dissip_invb20_out, 
      -f1.we*f1.delta_b_inv_out/f1.dissip_invb20_out, 
      -f2.we*f2.delta_b_inv_out/f2.dissip_invb20_out,
      -f3.we*f3.delta_b_inv_out/f3.dissip_invb20_out,
      r"$\mathrm{A}$", 1)

axs[3,1].set_ylim((0., 5))

axs[3,0].set_xlabel("Hour (h)", fontsize=myfontsize)
axs[3,1].set_xlabel("Hour (h)", fontsize=myfontsize)

mylabel = np.array((['a','b'],['c','d'], ['e','f'],['g','h']))

for i in range(4):
    for j in range (2):
        axs[i,j].text(0.95, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,
                      fontsize=myfontsize, fontweight='bold')

fig.savefig('Fig8_LES_tseries_diagnosis.png', dpi=fig.dpi, bbox_inches='tight')
```
