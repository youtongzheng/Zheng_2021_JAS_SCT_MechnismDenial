```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from matplotlib import cm
```

```python
f = xr.open_mfdataset('./*res.nc',
                      combine = 'by_coords',decode_cf = False)
```

```python
myfontsize = 12
```

```python
# defining a function for axe plotting
def pltbflx(iax, it, iz, ivar, ivarzi, ivarzb, ivarBIR):
    mylevels = np.linspace(-25, 25, 13)

    oax = iax.contourf(it,iz,ivar, cmap = cm.RdBu_r, levels = mylevels, extend = 'max')

    iax.plot(t, ivarzi, color = 'black',label=r"$\mathrm{z_i}$")
    iax.plot(t, ivarzb, color = 'black',label=r"$\mathrm{z_b}$", linestyle = '--')

    iax.set_ylabel('z (m)', fontsize=myfontsize)  # Add an x-label to the axes.
    iax.set_ylim([0., 1000])
    iax.legend(loc="upper left", fontsize=myfontsize, frameon=False, labelspacing = 0.2)

    iax0 = iax.twinx()  # instantiate a second axes that shares the same x-axis
    color = 'tab:blue'
    iax0.set_ylabel('BIR', color=color)  # we already handled the x-label with ax1
    iax0.plot(it, ivarBIR, color=color)
    iax0.tick_params(axis='y', labelcolor=color)
    iax0.set_ylim([-0.01, 0.16])
    
    return oax
```

```python
def plt2var(iax, it, ivar, ivar1, ivarname, ivarname1):
    iax.plot(t, ivar, color = 'black',label=ivarname)
    iax.set_ylabel(ivarname)
    
    iax0 = iax.twinx()  # instantiate a second axes that shares the same x-axis
    color = 'tab:red'
    iax0.set_ylabel(ivarname1, color=color)  # we already handled the x-label with ax1
    iax0.plot(it, ivar1, color=color)
    iax0.tick_params(axis='y', labelcolor=color)
    
    iax.grid(True)
```

```python
def pltcleb(iax, it, ivarEnt, ivarDiab, ivarRes, ivarStore, ivarRad, ivarPrec):
    iax.plot(it, ivarEnt,label='Ent')
    iax.plot(it, ivarDiab,label='Diab', color = 'black')
    iax.plot(it, ivarRes,label='Base')
    iax.plot(it, ivarStore,label='Stor')
    iax.plot(it, ivarRad,label='Rad', color = 'black', linestyle = '--', alpha=1)
    iax.plot(it, ivarPrec,label='Prec', color = 'black', linestyle = '-.', alpha=1)
    
    iax.set_ylabel(r"$\mathrm{W m^{-2}}$", fontsize=myfontsize)
    iax.set_xlabel("Hour (h)", fontsize=myfontsize)
    iax.set_ylim((-70, 70))
    iax.axhline(y=0., linewidth=1, color='black', linestyle='--')
    iax.legend(bbox_to_anchor=(1.,0.5), loc="center left", fontsize=0.9*myfontsize)

    iax.grid(True)
```

```python
zi = f['zi'].values[:,0]
BIR = f['BIR'].values[:,0]
indt = np.where(BIR < 0.15)[0]

z = f['z'].values
indz = np.where(zi < 1000.)[0]

fig, axs = plt.subplots(ncols = 1,
                        nrows = 4,
                        sharex = True, sharey = False,
                        figsize=(8/2.54, (14)/2.54), dpi = 300)
plt.tight_layout()

#
t = f['t_hr'].values[indt]
z = f['z'].values
var = f['sv_flux'].values[:, indt, 0]
varzi = f['zi'].values[indt, 0]
varzb = f['zb'].values[indt, 0]
varBIR = f['BIR'].values[indt, 0]

oax = pltbflx(axs[0], t, z, var, varzi, varzb, varBIR)

#
var = f['LHF'].values[indt, 0]
var1 = f['Bfx_mean'].values[indt, 0]

plt2var(axs[1], t, var, var1, r"$\mathrm{LHF \ (Wm^{-2})}$", r"$\mathrm{tot. buoy. \ (Wm^{-2})}$")

#
var = f['A'].values[indt, 0]
axs[2].plot(t, var, color = 'black',label=r"$\mathrm{A}$")
axs[2].set_ylabel(r"$\mathrm{A}$")
axs[2].set_ylim((1, 2.5))
axs[2].grid(True)

#
var = f['A'].values[indt, 0]
varEnt = f['Ent_CLEB'].values[indt, 0]
varDiab = f['Diab_CLEB'].values[indt, 0]
varRes = -f['Res_CLEB'].values[indt, 0]
varStore = f['Store_CLEB'].values[indt, 0]
varRad = f['Rad_CLEB'].values[indt, 0]
varPrec = f['Prec_CLEB'].values[indt, 0]

pltcleb(axs[3], t, varEnt, varDiab, varRes, varStore, varRad, varPrec)


cbar = fig.colorbar(oax, ax = axs, location = 'top',
                    fraction=0.07,
                    shrink=0.7, pad = 0.01, ticks=[-25, 0, 25])
cbar.set_label(r"$\mathrm{Buoy. flux \ (Wm^{-2})}$", fontsize=0.8*myfontsize)

mylabel = np.array(('a','b','c','d'))
for i in range(4):
    axs[i].text(0.01, 1.05, mylabel[i], transform=axs[i].transAxes,fontsize=myfontsize, fontweight='bold')
    axs[i].grid(True)
    
fig.savefig('Fig4_MLM_basic.png', dpi=fig.dpi, bbox_inches='tight')
```
